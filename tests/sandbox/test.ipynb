```python

financial_fields = ['ipa_funding',
 'ma_premium',
 'ma_risk_score',
 'mbr_with_rx_rebates',
 'partd_premium',
 'pcp_cap',
 'pcp_ffs',
 'plan_premium',
 'prof',
 'reinsurance',
 'risk_score_partd',
 'rx',
 'rx_rebates',
 'rx_with_rebates',
 'rx_without_rebates',
 'spec_cap'
]

for field in financial_fields:
    print(f"converting {field}")
    _df2 = _df2.withColumn(
        field,
        F.round(F.translate(field, "$,", "").cast("float"), 2)
    )

_df2.show()
```

```python
from pyspark.sql import functions as F

def convert_dollar_fields(df, dollar_fields):
    for field in dollar_fields:
        df = df.withColumn(
            field,
            F.round(F.translate(field, "$,", "").cast("float"), 2)
        )
    return df
```

```python
def rename_column(df):
    new_columns = [col.replace("rx", "pharma") for col in df.columns]
    renamed_df = df.toDF(*new_columns)
    return renamed_df

rename_column(_df2).show()
```
